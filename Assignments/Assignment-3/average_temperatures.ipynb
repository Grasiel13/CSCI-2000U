```python
#Sergio Majluf
#100558230

#importing all the needed documentation
from matplotlib import pyplot as plt
from matplotlib import patches as pch

import calendar
import numpy as np
import os 
%matplotlib inline
```

```python
files = os.listdir()
temperature_files = []
for file in files:
    if "temperature_2012.dat" in file:
        temperature_files.append(file)
```

```python
# add files to the plot.
plt.plotfile(temperature_files[0], color='blue', linewidth='2.5', names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[1], color='magenta', linewidth='2.5', newfig=False, names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[2], color='black', linewidth='2.5', newfig=False, names=calendar.month_abbr[1:13])

plt.xticks(np.arange(0,12), calendar.month_abbr[1:13])
line1 = pch.Patch(color='blue')
line2 = pch.Patch(color='magenta')
line3 = pch.Patch(color='black')
plt.legend((line1, line2, line3), ('Montreal', 'Toronto', 'Vancouver'), loc='upper right' )
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("2012 Average Temperatures (100558230)", fontsize=15)
# data ends at December, as it is only the 2012 year.
plt.axis([0,13,-10,25])

plt.show()
```
