```python
#getting the information from the file xy_points.dat
points = open("xy_points.dat", 'r')

x_list = []
y_list = []

for entry in points.readline().split(' '):
    x_list.append(float(entry))

for entry in points.readline().split(' '):
    y_list.append(float(entry))

x = np.array(x_list)
y = np.array(y_list)

x_sub0 = x[ (x < -2) ]
y_sub0 = y[ (x < -2) ]
x_sub1 = x_sub0[ (y_sub0 > -5) & (y_sub0 < 0) ]
y_sub1 = y_sub0[ (y_sub0 > -5) & (y_sub0 < 0) ]
x_rad = x[ ( ( ( (x)**2 + (y-5)**2 )**(1/2) ) <= 5 ) ]
y_rad = y[ ( ( ( (x)**2 + (y-5)**2 )**(1/2) ) <= 5 ) ]

plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x, y, color='red')
plt.scatter(x_sub1, y_sub1, color = 'blue')
plt.scatter(x_rad, y_rad, color='green')
plt.axis([-10,10,-10,10])
plt.title("Hilighted Points (100558230)", fontsize=14)

plt.show()
```
