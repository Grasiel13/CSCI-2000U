```python
#100558230
#Sergio Majluf

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

countries = ["USA", "CHN", "GBR", "RUS", "KOR", "GER", "FRA", "ITA", "HUN", "AUS", "OTHER"]
gold_medal = [46, 38, 29, 24, 13, 11, 11, 8, 8, 7, 107 ]
```

```python
cs = cm.Set2(np.arange(1, 255, 23))

plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
plt.title("2012 Olympic Gold Medals (100558230)", fontsize=16)
plt.pie(gold_medal, explode=[.15,.15,.15,.15,.15,.15,.15,.15,.15,.15,.15], labels=countries, 
        colors=cs, autopct='%1.1f%%', shadow=True, radius=0.95)

plt.show()
```
